# FTP - % of Bachelor's degree conferred to women (1970 - 2011) in the US

**Summary:**
* Women dominated majors (above 50%):
    * Health Professions, Public Administration, Education, Foreign Languages and English - best from 1970 to today
    * Psycology and Art and Performance - between 40% and 60% by 1970 and above 60% today
    * Communications and Journalism and Biology - slightly above 20% in 1970, above 50% today

### 2) Evolution of Bachelor degree where percentage of women graduates was above 40% in 1970

### 3) Evolution of Bachelor degree where percentage of women graduates was above 20% and below 40% in 1970

```python
between_20_40 = women_majors.loc[0, (women_majors.loc[0]>= 20) & (women_majors.loc[0]< 40)]
fte_between_20_40 = women_majors.plot(x = 'Year', y = between_20_40.index, figsize = (12,8), color = colours, legend = False)

fte_between_20_40.tick_params(axis = 'both', which = 'major', labelsize = 18)

fte_between_20_40.set_yticklabels(labels = [-10, '30  ', '35  ', '40  ','45  ',  '50   ','55  ' ,'60   ','65%'])

fte_between_20_40.axhline(y = 30, color = 'black', linewidth = 1.3, alpha = 0.7)

fte_between_20_40.set_xlim(left = 1969, right = 2011)

fte_between_20_40.xaxis.label.set_visible(False)

fte_between_20_40.text(x = 1965.8, y = 23, 
          s = '    pukkapad / Mariana                                                                        Source: National Center for Education Statistics  '   ,
          fontsize = 14, color = '#f0f0f0',  backgroundcolor = 'grey')

fte_between_20_40.text(x = 1966.65, y = 75,
           s = "Gender gap evolution for intermediary cases",
           fontsize = 26, weight = 'bold', alpha = 0.75)

fte_between_20_40.text(x = 1966.65, y = 70,
           s = "Percentage of Bachelors conferred to women from 1970 to 2011 in the US for\ncases where the percentage was above 20% and below 40% in 1970", 
            fontsize = 19, weight = 'bold', alpha = 0.75)

#Â Add labels
fte_between_20_40.text(x = 1988, y = 53, s = 'Biology', color = colours[0], weight = 'bold', rotation = 17,
             backgroundcolor = '#f0f0f0')
fte_between_20_40.text(x = 1973, y = 60, s = 'Communications and Journalism', color = colours[1], weight = 'bold', rotation = 57,
              backgroundcolor = '#f0f0f0')
fte_between_20_40.text(x = 2001, y = 48, s = 'Math and Statistics', color = colours[2], weight = 'bold', rotation = -17, 
               backgroundcolor = '#f0f0f0')
fte_between_20_40.text(x = 2001, y = 53, s = 'Social Sciences \nand History', color = colours[3], weight = 'bold', rotation = -14,
              backgroundcolor = '#f0f0f0')
```
